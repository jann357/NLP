```python

from collections import defaultdict, Counter


bigrams_list = list(nltk.bigrams(text))


bigram_freq = defaultdict(Counter)
unigram_freq = Counter(text)

for w1, w2 in bigrams_list:
    bigram_freq[w1][w2] += 1

vocab_size = len(unigram_freq)


def add_one_smoothing(bigram_freq, unigram_freq, vocab_size):
    smoothed_bigram_prob = {}
    for w1 in bigram_freq:
        smoothed_bigram_prob[w1] = {}
    for w2 in bigram_freq[w1]:
        smoothed_bigram_prob[w1][w2] = (bigram_freq[w1][w2] + 1) / (unigram_freq[w1] + vocab_size)
    return smoothed_bigram_prob

bigram_prob = add_one_smoothing(bigram_freq, unigram_freq, vocab_size)
```

```python
import nltk
from nltk.util import bigrams
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

text = open('names.txt').read()
#print(text[:400])

tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
sequences = tokenizer.texts_to_sequences([text])[0]

bigram_sequences = list(bigrams(sequences))
```

```python
X, y = zip(*bigram_sequences)
X = [[item] for item in X]
y = [[item] for item in y]
X = pad_sequences(X, maxlen=1)
y = pad_sequences(y, maxlen=1)
print(X)
print(y)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

model = Sequential()
model.add(Embedding(input_dim=len(tokenizer.word_index) + 1, output_dim=50, input_length=1))
model.add(LSTM(50))
model.add(Dense(len(tokenizer.word_index) + 1, activation='softmax'))


model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python

model.fit(X, y, epochs=500, verbose=2)
```

```python
import numpy as np

def generate_text(seed_text, next_words, model, tokenizer):
    for _ in range(next_words):
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences([token_list], maxlen=1, padding='pre')
        predicted = model.predict(token_list, verbose=0)
        predicted_word = tokenizer.index_word[np.argmax(predicted)]
        seed_text += " " + predicted_word
    return seed_text

seed_text = "around"
next_words = 2
print(generate_text(seed_text, next_words, model, tokenizer))
```
